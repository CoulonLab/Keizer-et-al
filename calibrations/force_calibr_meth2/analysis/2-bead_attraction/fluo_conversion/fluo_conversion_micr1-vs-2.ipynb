```python
from scipy import *
from numpy import *
from matplotlib import pyplot as plt
```

```python
(bead1_micr1_sig, bead1_micr1_bg, bead1_micr2_sig, bead1_micr2_bg,
 bead2_micr1_sig, bead2_micr1_bg, bead2_micr2_sig, bead2_micr2_bg,
 bead3_micr1_sig, bead3_micr1_bg, bead3_micr2_sig, bead3_micr2_bg) = loadtxt('fluo_conversion_micr1-vs-2.csv', delimiter=',', skiprows=1)[:,14]

micr1=r_[bead1_micr1_sig-bead1_micr1_bg,
         bead2_micr1_sig-bead2_micr1_bg,
         bead3_micr1_sig-bead3_micr1_bg,]
micr2=r_[bead1_micr2_sig-bead1_micr2_bg,
         bead2_micr2_sig-bead2_micr2_bg,
         bead3_micr2_sig-bead3_micr2_bg,]

fluoFactor_micr1_to_micr2=mean(micr2/micr1)

plt.figure()
plt.scatter(micr1,micr2);
x=r_[0,10e6]; plt.plot(x,x*fluoFactor_micr1_to_micr2)
plt.xlabel('ADU microscope 1'); plt.ylabel('ADU microscope 2'); 
plt.xlim(0,None); plt.ylim(0,None)
plt.show()
```

```python
print("Conversion factor: %.3f"%fluoFactor_micr1_to_micr2)
```
